# SVM   ---> classification if iris data

```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sn.load_dataset('iris')
df.head(2)
```

### Label Encoder ka  use krke speces ko    0 1 2 me convert kaernge

```python
from sklearn.preprocessing import LabelEncoder
le_species = LabelEncoder() 
```

### ek  new column (target) add karenge aur usme species ke no ko rakhenge  

```python
df['Target'] = le_species.fit_transform(df['species'])
```

```python
df
```

```python
df0 = df[:50]
df1 = df[51:100]
df2 = df[101 :]
```

```python
plt.figure(figsize = (7,5))
plt.xlabel('sepal_length')
plt.xlabel('sepal_width')

plt.scatter(df0['sepal_length'], df0['sepal_width'], label = 'setosa', marker = '*')
plt.scatter(df1['sepal_length'], df1['sepal_width'], label =  'versicolor')
plt.legend()
```

```python
plt.figure(figsize = (7,5))
plt.xlabel('petal_length')
plt.xlabel('petal_width')

plt.scatter(df0['petal_length'], df0['petal_width'], label = 'setosa', marker = '*')
plt.scatter(df1['petal_length'], df1['petal_width'], label =  'versicolor')
plt.legend()
```

#### X  &  Y decide Karenge 

```python
df.head(0)
```

```python
x = df.iloc[: , :4]
y = df.Target
```

## Data ko Train  &  Test  me Divide karenge

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2 )
```

```python
len(x_train)
```

```python
len(x_test)
```

## ab  SVM  algorithm ka use krenge

```python
from sklearn.svm import SVC
model = SVC()
```

### ab model ko Data dal kr train karenege

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
